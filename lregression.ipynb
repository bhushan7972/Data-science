```
import numpy as np

x=dataset['Years of Experience'].values
y=dataset['Salary'].values


def mean(values):
  return sum(values)/float(len(values))

x_mean=mean(x)
y_mean=mean(y)
  
  
numer=0
deno=0

for n in range(len(x)) :
  numer += (x[n]-x_mean)*(y[n]-y_mean)
  deno += (x[n]-x_mean)**2

m=numer/deno

b=y_mean-(m*x_mean)

print(m,b)  

plt.scatter(x,y,color='red')
plt.xlabel('year')
plt.ylabel('salary')

xmax=np.max(x)+1
xmin=np.min(x)-1

X=np.linspace(xmax,xmin,10)
X=np.
Y=m*X+b

plt.plot(X,Y,color='green')
```

```
def predict(x):
  return m*x+b

def checkerror(y,y_pred):
  return abs(y-y_pred)  


y_pred=predict(4.4)  
print("predicted value:",y_pred)
print("actual value:",46025)
print("error is",checkerror(46025,y_pred))
```
